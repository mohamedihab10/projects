```python
lr = LinearRegression()
```

```python
errors = []
for train , test in splits:
    x_train1, y_train1 = X[train] , y[train]
    x_val1 , y_val1 = X[test],y[test]
    lr.fit(x_train1 ,y_train1)
    pred =lr.predict(x_val1)
    errors.append(mean_absolute_error(pred ,y_val1))
```
